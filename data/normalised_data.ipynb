```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import joblib
```

```python
# 1. Load the original raw data
original_df = pd.read_csv('smart_water_distribution_data.csv')
```

```python
# 2. Make a copy so we do NOT permanently change the original
scaled_df = original_df.copy()
```

```python
# 3. Define the columns that actually need to be scaled (numbers, not timestamps)
columns_to_scale = [
    'Hour_of_Day', 'Day_of_Week', 'Month', 
    'Temperature_C', 'Rainfall_mm', 
    'Lag_1_Demand', 'Lag_24_Demand', 'Current_Demand'
]
```

```python
# 4. Initialize the Scaler
scaler = MinMaxScaler()
```

```python
# 5. Fit the scaler to the data and transform it
scaled_df[columns_to_scale] = scaler.fit_transform(scaled_df[columns_to_scale])
```

```python
# 6. Save the newly scaled data into a DIFFERENT CSV file
scaled_df.to_csv('scaled_smart_water_distribution_data.csv', index=False)
```

```python
# 7. IMPORTANT FOR YOUR BACKEND ENGINEER: Save the scaler!
joblib.dump(scaler, 'scaler.pkl')

print("✅ Data scaled and saved to 'scaled_smart_water_distribution_data.csv'")
```
